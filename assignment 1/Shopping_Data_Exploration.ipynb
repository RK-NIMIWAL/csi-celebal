## Step 3: Handle Missing Values

### 3a · Identify missing values

```python
missing = df.isnull().sum()
missing_pct = (df.isnull().mean() * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print(missing_df[missing_df['Missing Count'] > 0])
```

### 3b · Fill / Drop missing values

```python
# Fix numeric columns (coerce non-numeric to NaN)
df['price']    = pd.to_numeric(df['price'],    errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')

# Fill numeric missing values with column median
df['price']    = df['price'].fillna(df['price'].median())
df['quantity'] = df['quantity'].fillna(df['quantity'].median())

# Fill categorical missing values with column mode
df['payment_method'] = df['payment_method'].fillna(df['payment_method'].mode()[0])
df['city']           = df['city'].fillna(df['city'].mode()[0])

print('Missing values after handling:')
print(df.isnull().sum())
```

## Step 4: Basic Operations — Filter Rows & Select Columns

### 4a · Filter: Orders where price > 30

```python
high_value = df[df['price'] > 30]
print(f'High-value orders (price > ₹30): {len(high_value)}')
high_value.head()
```

### 4b · Filter: Electronics category only

```python
electronics = df[df['category'] == 'Electronics']
print(f'Electronics orders: {len(electronics)}')
electronics.head()
```

### 4c · Select specific columns

```python
selected = df[['order_id', 'product_name', 'category', 'price', 'quantity']]
selected.head()
```

### 4d · Filter: UPI payments with quantity ≥ 3

```python
upi_bulk = df[(df['payment_method'] == 'UPI') & (df['quantity'] >= 3)]
print(f'UPI + quantity ≥ 3 orders: {len(upi_bulk)}')
upi_bulk.head()
```

## Step 5: Remove Duplicates

```python
print(f'Rows before removing duplicates: {df.shape[0]}')
print(f'Duplicate rows found: {df.duplicated().sum()}')

df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

print(f'Rows after removing duplicates: {df.shape[0]}')
```

## Step 6: Create Derived Column — `total_amount`

```python
df['total_amount'] = df['price'] * df['quantity']
df['total_amount'] = df['total_amount'].round(2)

print('New column added: total_amount = price × quantity')
df[['order_id', 'product_name', 'price', 'quantity', 'total_amount']].head(10)
```

### 6a · Summary stats of total_amount by category

```python
summary = df.groupby('category')['total_amount'].agg(['sum','mean','count'])
summary.columns = ['Total Revenue', 'Avg Order Value', 'Order Count']
summary['Total Revenue'] = summary['Total Revenue'].round(2)
summary['Avg Order Value'] = summary['Avg Order Value'].round(2)
summary.sort_values('Total Revenue', ascending=False)
```

## Step 7: Save the Cleaned Dataset

```python
df.to_csv('shopping_cleaned.csv', index=False)
print(f'Cleaned dataset saved as: shopping_cleaned.csv')
print(f'Final shape: {df.shape[0]} rows × {df.shape[1]} columns')
print(f'Columns: {df.columns.tolist()}')
```
